A Python program to manage student records using file handling. Your program should:

1. Ask the user to enter student details (Name, Roll Number, Marks).
2. Save the details into a text file (students.txt).
3. Read and display all student records from the file.
4. Append new student records without overwriting existing ones.
5. Handle cases where the file does not exist (show proper error message).
6. Allow the user to search for a student by roll number.

```python
import os

#file where students records will be stored
filename = "students.txt"

#fn to add a new student record
def add_student():
    name = input("Enter student name: ")
    roll = input("Enter roll number: ")
    marks = input("Enter marks: ")
    try:
        #"a" append mode(creates file of not exists)
        with open(filename, "a") as file:
            file.write(f"{name},{roll},{marks}\n")
        print("Student record added successfully.")
    except Exception as e:
        print("Error adding student record:", e)

#fn to read and display all student records
def read_students():
    if not os.path.exists(filename):
        print("No student records found.")
        return
    try:
        with open(filename, "r") as f:
            data + f.readlines()
            if not data:
                print("no student records available.")
            else:
                print("---- Student Records ----")
                for line in data:
                    print(line.strip())#remove \n while printing
    except Exception as e:
        print("Error reading file:", e)

#fn to search a student record by roll number
def search_student():
    roll = input("enter roll number to search:")
    found = False
    try:
        with open(filename, "r") as f:
            for line in f:
                if line.startswith(roll + ","): #check roll number at start of line
                    print("Student found:", line.strip())
                    found = True
                    break
        if not found:
            print("Student with roll number", roll, "not found.")
    except FileNotFoundError:
        print("file not found! please add student records first.")
    except Exception as e:
        print("Error searching student record:", e)

#main menu loop
while True:
    print("\n--- Student Record Management ---")
    print("1. Add Student")
    print("2. View All Students")
    print("3. Search Student by Roll Number")
    print("4. Exit")
    choice = input("Enter your choice (1-4): ")
    
    if choice == '1':
        add_student()
    elif choice == '2':
        read_students()
    elif choice == '3':
        search_student()
    elif choice == '4':
        print("Exiting program.")
        break
    else:
        print("Invalid choice! Please enter a number between 1 and 4.")
```
